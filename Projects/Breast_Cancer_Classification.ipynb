```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets 
from sklearn.model_selection import train_test_split
```

```python
# Data Collection and Preprocessing 
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the dataset as dataframe 
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.info()
```

```python
# checking for missin values 
data_frame.isnull().sum()
```

```python
data_frame.describe()
```

```python
# checking the distribution of target variable 
data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
# Separating the features and targets 

x = data_frame.drop('label', axis=1)
y = data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

```python
# separating for the test and train datasets 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
# Scaling the data 
from sklearn.preprocessing import StandardScaler 
std = StandardScaler()

x_train_std = std.fit_transform(x_train)
x_test_std = std.transform(x_test)
```

```python
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

#### Building the Neural Network

```python
# importing tf and keras 
import tensorflow as tf  

# for initalization of weights and biases
tf.random.set_seed(3)

from tensorflow import keras
```

```python
# setting up the layers of NN 
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiliing the NN 
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# training the NN 

history = model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
```

```python
# Visualization of the accuracy and loss 
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['Training Data', 'Validation Data'], loc='lower right')
```

```python
# Visualization of the accuracy and loss 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('Epoch')

plt.legend(['Training Data', 'Validation Data'], loc='upper right')
```

```python
# Checking the accuracy on test dataset 

loss, accuracy = model.evaluate(x_test_std,y_test)
print(accuracy)
```

```python
print(x_test_std.shape)
print(x_test_std[0])
```

```python
y_pred = model.predict(x_test_std)
```

```python
print(y_pred.shape)
print(y_pred[0])
```

model.predict() gives the prediction probability of each class for the datapoint

```python
# Converting the prediction probability to class labels 

y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
